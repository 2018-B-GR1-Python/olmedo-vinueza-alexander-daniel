```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.histogram([1,2,1],bins=[0,1,2,3])
mu, sigma = 2, 0.5
```

```python
data = np.random.normal(mu,sigma,10000)
```

```python
data
```

```python
(n, bin_edges) = np.histogram(data,bins=50)
plt.plot(bin_edges[1:],n)
plt.show()
```
